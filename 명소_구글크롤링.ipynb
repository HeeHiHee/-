```python
def getreview(keyword,driver):
    url=('https://www.google.co.kr/maps/search/'+keyword+'/@35.8698753,128.5880298,15z/data=!3m1!4b1?hl=ko')
    driver.get(url)
    add=driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[7]/div[1]/button/div[1]/div[2]/div[1]").text
    
    
    
    
    driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[2]/div[1]/div[1]/div[2]/div/div[1]/span[1]/span/span/span[2]/span[1]/button").click()
    
    driver.implicitly_wait(10)
    
    num=driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[2]/div[2]/div/div[2]/div[2]")
    num=num.text
    string =num
    numbers = re.sub(r'[^0-9]', '', string)
    numbers=int(numbers)
    
    #창 뜰때까지 기다리기
    driver.implicitly_wait(10)
    
    #세부창 스크롤 class name으로 받아오기
    itemlist = driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[2]")
    
    starp=[]
    review=[]
    date=[]
    
    driver.implicitly_wait(10)
    
    

    for i in range(numbers):
        try:
            star=driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[2]/div[9]/div["+str(1+(3*i))+"]/div/div[3]/div[3]/div[1]/span[2]")
            reviews=driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[2]/div[9]/div["+str(1+(3*i))+"]/div/div[3]/div[3]/div[2]/span[2]")
            dates=driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[2]/div[9]/div["+str(1+(3*i))+"]/div/div[3]/div[3]/div[1]/span[3]")

            driver.execute_script("arguments[0].scrollBy(0, 5000)", itemlist)

            starp.append(star.get_attribute('aria-label'))
            review.append(reviews.text)
            date.append(dates.text)

            driver.implicitly_wait(10) 
            
        except NoSuchElementException:
            print("에러를 무시하며 작업을 계속합니다.")

    df=pd.DataFrame({'starp':starp,
                     'review':review,
                     'date':date})
  

    df['add']=add
    df['위치']=keyword
    df['리뷰수']=numbers
    
    return df
```

```python
def crawlreview(keyword):
    driver = webdriver.Chrome('C:\\chromedriver.exe')
    return getreview(keyword,driver)
```

## datetime계산

```python
#계산 순서
# 1. 구글에서는 "몇 x시간 전 이렇게 표시된다,"
# 2. 초인지 분인지 일인지 년인지에 따라 if문 구성하고
# 3. 한글을 숫자로 구분하고
# 4. 날짜에서 해당 숫자에 해당하는 단위로 빼준다.  relativedelta함수 사용




for i in range(len(df)):
    if '초' in df['date'][i]:
        num=re.sub(r'[^0-9]', '',df['date'][i])
        df['date'][i]=now-relativedelta(seconds=int(num))
    elif '분' in df['date'][i]:
        num=re.sub(r'[^0-9]', '',df['date'][i])
        df['date'][i]=now-relativedelta(minutes=int(num))
    elif '시간' in df['date'][i]:
        num=re.sub(r'[^0-9]', '',df['date'][i])
        df['date'][i]=now-relativedelta(hours=int(num))
    elif '일' in df['date'][i]:
        num=re.sub(r'[^0-9]', '',df['date'][i])
        df['date'][i]=now-relativedelta(days=int(num))
    elif '주' in df['date'][i]:
        num=re.sub(r'[^0-9]', '',df['date'][i])
        df['date'][i]=now-relativedelta(weeks=int(num))
    elif '달' in df['date'][i]:
        num=re.sub(r'[^0-9]', '',df['date'][i])
        df['date'][i]=now-relativedelta(months=int(num))
    elif '일' in df['date'][i]:
        num=re.sub(r'[^0-9]', '',df['date'][i])
        df['date'][i]=now-relativedelta(days=int(num))
    elif '년' in df['date'][i]:
        num=re.sub(r'[^0-9]', '',df['date'][i])
        df['date'][i]=now-relativedelta(years=int(num))  
```

```python
def get_date(df):
    for i in range(len(df)):
        if '초' in df['date'][i]:
            num=re.sub(r'[^0-9]', '',df['date'][i])
            df['date'][i]=now-relativedelta(seconds=int(num))
        elif '분' in df['date'][i]:
            num=re.sub(r'[^0-9]', '',df['date'][i])
            df['date'][i]=now-relativedelta(minutes=int(num))
        elif '시간' in df['date'][i]:
            num=re.sub(r'[^0-9]', '',df['date'][i])
            df['date'][i]=now-relativedelta(hours=int(num))
        elif '일' in df['date'][i]:
            num=re.sub(r'[^0-9]', '',df['date'][i])
            df['date'][i]=now-relativedelta(days=int(num))
        elif '주' in df['date'][i]:
            num=re.sub(r'[^0-9]', '',df['date'][i])
            df['date'][i]=now-relativedelta(weeks=int(num))
        elif '달' in df['date'][i]:
            num=re.sub(r'[^0-9]', '',df['date'][i])
            df['date'][i]=now-relativedelta(months=int(num))
        elif '일' in df['date'][i]:
            num=re.sub(r'[^0-9]', '',df['date'][i])
            df['date'][i]=now-relativedelta(days=int(num))
        elif '년' in df['date'][i]:
            num=re.sub(r'[^0-9]', '',df['date'][i])
            df['date'][i]=now-relativedelta(years=int(num))  

    return df
```
